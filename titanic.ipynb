```python
df_train = pd.read_csv('train.csv')
```

```python
# Visualizando a base de teste
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
df_test.info()
```

```python
# Verificando os valores vazios
df_test.isnull().sum().sort_values(ascending=False).head(5)
```

```python
# Retornando os valores vazios do treino
df_train.isnull().sum().sort_values(ascending=False).head(5)
```

```python
# Retorna o shape da base
df_train.shape
```

```python
# Verificando a cardinalidade dos dados
df_train.nunique().sort_values(ascending=False)
```

```python
# Eliminando da base de treino colunas com alta cardinalidade
df_train = df_train.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Eliminando também para a base de teste
df_test = df_test.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Verificando a média das idades
df_train.Age.mean()
```

```python
# Podemos substituir os valores faltantes das idades plela idade média da base
df_train.loc[df_train.Age.isnull(), 'Age'] = df_train.Age.mean()
```

```python
# Faezendo a mesma coisa para a base de teste
df_test.loc[df_test.Age.isnull(), 'Age'] = df_test.Age.mean()
```

```python
# Verificando os valores na coluna Embarked
df_train.Embarked.value_counts()
```

```python
# Selecionando a moda dessa coluna
df_train.Embarked.mode()[0]
```

```python
# Substituindo os valores vazios na coluna Embarked pela moda
df_train.loc[df_train.Embarked.isnull(), 'Embarked'] = df_train.Embarked.mode()[0]
```

```python
# Substituindo o único valor nulo da coluna Fare da base de teste pela sua média
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test.Fare.mean()
```

```python
# Verificando novamente se ainda existe algum valor nulo na base de treino
df_train.isnull().sum().sort_values(ascending=False).head(5)
```

```python
# Verificando novamente se ainda existe algum valor nulo na base de teste
df_test.isnull().sum().sort_values(ascending=False).head(5)
```

```python
# Transformando as variáveis categóricas úteis em numéricas
def transforma_sexo(valor):
    if valor == 'male':
        return 1
    else:
        return 0

def transforma_embarque(valor):
    if valor == 'S':
        return 0
    elif valor == 'C':
        return 1
    else:
        return 2

df_train.Sex = df_train.Sex.apply(transforma_sexo)
df_train.Embarked = df_train.Embarked.apply(transforma_embarque)
df_test.Sex = df_test.Sex.apply(transforma_sexo)
df_test.Embarked = df_test.Embarked.apply(transforma_embarque)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
# Após manipular os dados, podemos escolher os modelos para classificar esses dados: Árvore de classificação, classificação dos vizinhos mais próimos, regressão logística, etc
```

```python
# Importando o train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Separando o X e o y
X = df_train.drop(['PassengerId', 'Survived'], axis=1)
y = df_train.Survived
```

```python
# Separando os dados de treino e teste
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Utilizando primeiramente a árvore de classificação
from sklearn import tree
```

```python
# Criando o classificador
clf_ac = tree.DecisionTreeClassifier(random_state=42)
```

```python
# Fazendo o fit com os dados
clf_ac = clf_ac.fit(X_train, y_train)
```

```python
# Fazendo a previsão com os dados
y_pred_ac = clf_ac.predict(X_val)
```

```python
# Utilizando o KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Criando o classificador
clf_knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Fazendo o fit com os dados
clf_knn = clf_knn.fit(X_train, y_train)
```

```python
# Fazendo a previsâo
y_pred_knn = clf_knn.predict(X_val)
```

```python
# Utilizando a regressão logística
from sklearn.linear_model import LogisticRegression
```

```python
# Criando o classificador
clf_lr = LogisticRegression(random_state=42)
```

```python
# Fazendo o fit com os dados
clf_lr = clf_lr.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_lr = clf_lr.predict(X_val)
```

```python
# Para finalizar, vamos analisar os três modelos utilizados através da acurácia
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Para a árvore
accuracy_score(y_val, y_pred_ac)
```

```python
# Para o KNN
accuracy_score(y_val, y_pred_knn)
```

```python
# Para a regressão
accuracy_score(y_val, y_pred_lr)
```

```python
# Avaliando agora a matriz de confusão
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# Para a árvore
confusion_matrix(y_val, y_pred_ac)
```

```python
# Para o KNN
confusion_matrix(y_val, y_pred_knn)
```

```python
# Para a regressão
confusion_matrix(y_val, y_pred_lr)
```

Agora, faremos a normalização dos dados para ver se alguma acurácia se dá diferente

```python
# Importando o MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Selecionar só as colunas numéricas
colunas_numericas = df_train.drop(['PassengerId', 'Survived'], axis=1).select_dtypes(include=['int64', 'float64']).columns

# Aplicar normalização
df_train[colunas_numericas] = scaler.fit_transform(df_train[colunas_numericas])

df_train.head()
```

```python
# Fazendo o mesmo trabalho feito anteriormente, mas agora com os dados normalizados
X = df_train.drop(['PassengerId', 'Survived'], axis=1)
y = df_train.Survived
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
clf_ac_norm = tree.DecisionTreeClassifier(random_state=42)
clf_ac_norm = clf_ac_norm.fit(X_train, y_train)
y_pred_ac_norm = clf_ac_norm.predict(X_val)
```

```python
clf_knn_norm = KNeighborsClassifier(n_neighbors=3)
clf_knn_norm = clf_knn_norm.fit(X_train, y_train)
y_pred_knn_norm = clf_knn_norm.predict(X_val)
```

```python
clf_lr_norm = LogisticRegression(random_state=42)
clf_lr_norm = clf_lr_norm.fit(X_train, y_train)
y_pred_lr_norm = clf_lr_norm.predict(X_val)
```

```python
accuracy_score(y_val, y_pred_ac_norm)
```

```python
accuracy_score(y_val, y_pred_knn_norm)
```

```python
accuracy_score(y_val, y_pred_lr_norm)
```

Após normalizar os dados, vimos que o único modelo que sofreu um efeito positivo da normalização foi o KNN, que passou de uma acurácia de 71,5% para 78,3%. Porém, a regressão logística, tanto com normalização quanto sem, mostrou a melhor acurácia, com 81%. Então, ela será a escolhida para tratar os dados de teste
